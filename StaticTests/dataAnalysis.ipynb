```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Reading time analysis

```python
df = pd.read_csv('./test_batery8-3.csv')
```

```python
df['time_diff'] = df['Time'].diff()

number_of_readings = len(df['time_diff'])
total_time = df["Time"].iloc[-1] - df["Time"].iloc[0]
max_diff = df['time_diff'].max()
min_diff = df['time_diff'].min()
mean_diff = df['time_diff'].mean()
```

```python
print("-=-=- General Statistics -=-=-")
print(f"Max: {max_diff} ms")
print(f"Min: {min_diff} ms")
print(f"Amplitude: {(max_diff - min_diff):.2f} ms")
print(f"Mean: {(mean_diff):.2f} ms")
print(f"Total time: {(total_time/1000):.2f} s")
print(f"Average readings per second: {(number_of_readings / total_time*1000):.2f}")

# Histogram of time between each reading
plt.plot(df['time_diff'])
plt.title('Time between readings')
plt.xlabel('Reading number')
plt.ylabel('Time difference (ms)')
plt.show()
```

**Notes:**
- The system seems to register less data when the motor is at it's max power
- The time difference varies a lot, but the reasons aren't clear
- Good news: even on it's peak, the time difference isn't long

## Motor performance analysis

```python
df = pd.read_csv('./motorPerformance.csv')
```

```python
# Bar charts
plt.bar(df['embedded'], df['Vm'])
plt.title('Voltage')
plt.xlabel('Embedded')
plt.ylabel('Vm')

for i, v in enumerate(df['Vm']):
    plt.text(i, v, str(v), ha='center', va='bottom')

plt.show()

plt.bar(df['embedded'], df['Wp'])
plt.title('Power')
plt.xlabel('Embedded')
plt.ylabel('Wp')

for i, v in enumerate(df['Wp']):
    plt.text(i, v, str(v), ha='center', va='bottom')

plt.show()

plt.bar(df['embedded'], df['Ah'])
plt.title('Charge')
plt.xlabel('Embedded')
plt.ylabel('Ah')

for i, v in enumerate(df['Ah']):
    plt.text(i, v, str(v), ha='center', va='bottom')

plt.show()

plt.bar(df['embedded'], df['Wh'])
plt.title('Energy')
plt.xlabel('Embedded')
plt.ylabel('Wh')

for i, v in enumerate(df['Wh']):
    plt.text(i, v, str(v), ha='center', va='bottom')

plt.show()

plt.bar(df['embedded'], df['Ap'])
plt.title('Current')
plt.xlabel('Embedded')
plt.ylabel('Ap')

for i, v in enumerate(df['Ap']):
    plt.text(i, v, str(v), ha='center', va='bottom')

plt.show()
```

**Notes:**
- It's relevant to know that the first test was with the embedded off and the battery was not charged between tests
- Good news: the peak power stayed the same
